```python
my_map = folium.Map(
    location = [13.133032434766733, 16.103938729508073],
    zoom_start=2
)
for _, city in cities.iterrows():
    folium.Marker(
        location = [city['latitude'], city['longitude']],
        popup = city['name'],
        tooltip = city['name'],
        icon = folium.Icon(color=city['color'],prefix='fa',icon='circle')
    ).add_to(my_map)
my_map
```

```python
def select_marker_color(row):
    if row['lived'] == 'yes':
        return 'pink'
    elif row['lived'] == 'no' and row['visited'] == 'yes':
        return 'purple'
    return 'blue'
```

```python
cities['color'] = cities.apply(select_marker_color,axis=1)
cities
```
